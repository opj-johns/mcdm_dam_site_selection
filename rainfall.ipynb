## Rainfall Data from NASA Power API. 
## For specific locations.
## From 2010 to 2024

```python
import requests
import time
import pandas as pd
```

```python
monthly_url = "https://power.larc.nasa.gov/api/temporal/monthly/point"
```

```python
df = pd.read_csv('dams.csv')
```

```python
commune_rain_data = {}
```

```python
for index in range(0, len(df)):
    monthly_params = {
    "parameters": "PRECTOTCORR",
    "community": "AG",
    "longitude": df.longitude[index],
    "latitude": df.latitude[index],
    "start": "2010",
    "end": "2024",
    "format": "JSON"
    }

    monthly_response = requests.get(monthly_url, params=monthly_params)
    # wait for 5 seconds
    if monthly_response.status_code == 200:
        monthly_data = monthly_response.json()['properties']['parameter']['PRECTOTCORR']
        count = 0
        sum = 0
        yearly_rainfall = []
        for val in monthly_data.items():
            if count <= 12:
                sum += val[1]
                count+=1
                continue
            yearly_rainfall.append(sum)
            count = 0
            sum = 0
        commune_rain_data[df.nom_barrag[index]] = yearly_rainfall
    else:
        print("error fetching data")
        commune_rain_data[df.nom_barrag[index]] = -999
```

```python
commune_rain_data
```

```python
import csv

# Define years from 2010 to 2024
years = list(range(2010, 2025))

# Write to CSV
with open("rain_data.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    
    # Write header
    writer.writerow(["name"] + years)
    
    # Write rows
    for name, values in commune_rain_data.items():
        writer.writerow([name] + values)
```
